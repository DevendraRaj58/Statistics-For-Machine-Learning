```python
train=pd.read_csv('/kaggle/input/competitions/titanic/train.csv').drop(columns=['Survived'])
test=pd.read_csv('/kaggle/input/competitions/titanic/test.csv')
```

```python
final_df=pd.concat([train,test]).sample(1309)
```

```python
pop=final_df['Age'].dropna()
```

```python
sample_age=pop.sample(25).values
```

```python
sample_age
```

```python
#NUll Hypothesis H0: Population mean age is 35
#Alternative Hypothesis Ha: Population mean age is not 35
```

```python
#Check for normality using shapiro
from scipy.stats import shapiro

shapiro_age=shapiro(sample_age)
print(shapiro_age)
```

pvalue given by shapiro function is greater than 0.05
this signifies that the sample data is normally distributed

```python
pop_mean=35
import scipy.stats as stats
tstatistic,p_value=stats.ttest_1samp(sample_age,pop_mean)
print("t_statistic",tstatistic)
print("p_value:",p_value/2)
```

```python
significance_level=0.05
if p_value<significance_level:
    print("Reject Null Hypothesis")
else:
    print("Cannot reject Null Hypothesis")
```

Our Null Hyptothesis was population mean is equal to 35 but our test proves that this is false
