```python
# Input the data as lists
desktop_users = [12, 15, 18, 16, 20, 17, 14, 22, 19, 21, 23, 18, 25, 17, 16, 24, 20, 19, 22, 18, 15, 14, 23, 16, 12, 21, 19, 17, 20, 14]
mobile_users = [10, 12, 14, 13, 16, 15, 11, 17, 14, 16, 18, 14, 20, 15, 14, 19, 16, 15, 17, 14, 12, 11, 18, 15, 10, 16, 15, 13, 16, 11]
```

```python
from scipy.stats import shapiro

shapiro_mobile=shapiro(mobile_users)
shapiro_desktop=shapiro(desktop_users)

print("Mobile Users:",shapiro_mobile)
print("Desktop Users:",shapiro_desktop)
```

**P-Value of both samples is greater than significance level hence both samples are nomrally distributed**

```python
#Using levene test to check for homoscedasity(approximately equal standard deviation for both samples)

from scipy.stats import levene
levene_test=levene(desktop_users,mobile_users)
print(levene_test)
```

**P-Value through levene test is found to be higher than significance level(0.05) which means test for homoscedascity is true**

```python
#This gives the area under the extreme region both to the left and right side
#Way 2 using cdf function
# from scipy.stats import t
# t_value=-5.25
# dof=58
# cdf_value=t.cdf(t_value,dof)
# print(cdf_value*2)
#Way 2 using function
import scipy.stats as st
statistic,p_value=st.ttest_ind(desktop_users,mobile_users)
print(statistic)
print(p_value)
```

**The cdf value is very close to zero i.e less than significance level. So we will Reject The Null Hypothesis**
