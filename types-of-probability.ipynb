```python
df_1=pd.read_csv(r'/kaggle/input/competitions/titanic/train.csv')
df_2=pd.read_csv(r'/kaggle/input/competitions/titanic/test.csv')
```

```python
df=pd.concat([df_1,df_2])
```

```python
df
```

```python
#This represents the Joint Probability
pd.crosstab(df['Survived'],df['Pclass'],normalize='all')
```

```python
#This represents the Marginal Probability
pd.crosstab(df['Survived'],df['Pclass'],normalize='all',margins=True)
```

```python
# Understanding and implementing conditional probability
# Coinditional Probability is the probability of an event occuring given other event has already occured

pd.crosstab(df['Survived'],df['Pclass'],normalize='columns')
#Above code gives the conditional probability of survival provided they are in a given Pclass
```

```python
'''
Below code line is just the opposite of the previous cell code.
Which means now the already occured event is being in pclass and the next event is survival.
This code gives us the probability of a death/survival for particular pclass
'''
pd.crosstab(df['Survived'],df['Pclass'],normalize='index')
```
