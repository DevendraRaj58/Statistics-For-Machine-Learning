```python
#Hypothesis Test: Average male age is significantly greater than average female age for titanic passengers
#H0: Mean age is similar for male and female
#Ha: Mean age is significantly different for male and female
#Significance level alpha= 0.05
alpha=0.05
```

```python
train=pd.read_csv('/kaggle/input/competitions/titanic/train.csv')
test=pd.read_csv('/kaggle/input/competitions/titanic/test.csv')
```

```python
final_df=pd.concat([train,test]).sample(1309)
```

```python
pop_male=final_df[final_df['Sex']=='male']['Age'].dropna()
pop_female=final_df[final_df['Sex']=='female']['Age'].dropna()
```

```python
sample_male=pop_male.sample(25)
sample_female=pop_female.sample(25)
```

```python
#Normality check for both samples using Shapiro-Wilk Test
from scipy.stats import shapiro
shapiro_male=shapiro(sample_male)
shapiro_female=shapiro(sample_female)
print("Shapiro Test for male:",shapiro_male)
print("Shapiro Test for female:",shapiro_female)
```

**Since the P-value for both male and female age sample is greater than significance level (0.05), the sample data is normally distributed**

```python
#Check for homoscedascity using Levene test
from scipy.stats import levene
levene_test=levene(sample_male,sample_female)
print(levene_test)
```

**Since the p-value given by levene test is greater than significance level , test for homoscedasity is true i.e Standard deviation is approximately equal of both the sample data**

```python
#Using ttest_ind function to calculate p-value for both samples
import scipy.stats as st
statistic,p_value=st.ttest_ind(sample_male,sample_female)
print("Statistic:",statistic)
print("P-Value:",p_value)
```

```python
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Cannot reject null Hypothesis")
```
