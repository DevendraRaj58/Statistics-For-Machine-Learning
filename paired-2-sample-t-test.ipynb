# **This Notebook Explores the Paired 2-Sample T-test for hypothesis testing.**
### The test is conducted on a basic data of weight before and after a weight reduction program
### Null Hypothesis (H0): Weight remains similar before and after the program
### Alternate Hypothesis (Ha): Weight reduces after the program

```python
weight_before=np.array([78, 93, 81, 67, 88, 76, 74, 91, 69, 88, 77, 81, 80, 79, 88])
weight_after=np.array([80, 92, 75, 68, 85, 78, 73, 90, 70, 88, 76, 84, 82, 77, 91])
```

```python
differences=weight_before-weight_after
```

```python
mean_difference=np.mean(differences)
std_dev=np.std(differences,ddof=1)
print(std_dev)
```

```python
#Shapiro test to check for normality
from scipy.stats import shapiro
shapiro_test=shapiro(differences)
print(shapiro_test)
#Since the p value for the test is greater than significance level (0.05) normality test is true
```

```python
n=len(differences)
t_value=mean_difference/(std_dev/np.sqrt(n))
print(t_value)
df=n-1
```

```python
from scipy.stats import t
alpha=0.05
p_value=t.cdf(t_value,df)
print(p_value)
```

**P-value is greater than the significance level due to which we cannot reject null hypothesis**

**This implies weight before and after the program remains similar but here is the CATCH!!**

**From data it is clear that there is a difference between the weights yet our hypothesis proved similar weight. Hence it means that we are rejecting the alternative hypothesis but at the same time not rejecting null hypothesis does not mean that null hypothesis is correct.This just simply means we couldn't gather enough evidence to prove that.**
